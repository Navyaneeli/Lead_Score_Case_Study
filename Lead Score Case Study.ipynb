 #Step 3: Data Preparation

```python
# handling missing values
100*(df.isnull().sum()/len(df))
```

```python
#data preparation
# List of variables to map

varlist =  ['Specialization','How did you hear about X Education','Lead Profile','City','Do Not Email', 'Do Not Call', 'Search', 'Magazine', 'Newspaper Article','X Education Forums','Newspaper','Digital Advertisement','Through Recommendations','Receive More Updates About Our Courses','Update me on Supply Chain Content','Get updates on DM Content','A free copy of Mastering The Interview']

# Defining the map function
def binary_map(x):
    return x.map({'Yes': 1, "No": 0,"Select": " " })

# Applying the function to the housing list
df[varlist] = df[varlist].apply(binary_map)
```

```python
100*(df.isnull().sum()/len(df))
```

```python
#deleting the columns having morethan 40% of missing values
missing_vaues_column_list=df[df.columns[100*df.isnull().sum()/len(df)>40]]
df.drop(missing_vaues_column_list, axis=1, inplace=True)
```

```python
100*(df.isnull().sum()/len(df))
```

```python
df.shape
```

```python
#dropping the columns which are not needed for the analysis
df.drop(["Country","Lead Number","Tags","I agree to pay the amount through cheque"],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.rename(columns={'What is your current occupation': 'occupation', 'What matters most to you in choosing a course': 'choosing_a_course','Lead Source':"Lead_Source",'Page Views Per Visit':"Page_Views_Per_Visit","Last Activity":"Last_Activity"}, inplace=True)
```

```python
#dropping the null rows of occupation since the count of missing values is more
df=df[~df.occupation.isnull()].copy()
```

```python
#dropping the null rows of choosing_a_course since the count of missing values is more
df=df[~df.choosing_a_course.isnull()].copy()
```

```python
100*(df.isnull().sum()/len(df))
```

```python
#find the mode of Lead Source in df
lead_mode=df.Lead_Source.mode()[0]
lead_mode
```

```python
# fill the missing values with mode value of occupation in df.
df.Lead_Source.fillna(lead_mode,inplace=True)
```

```python
#find the mode of Last_Activity in df
last_mode=df.Last_Activity.mode()[0]
last_mode
# fill the missing values with mode value of occupation in df.
df.Last_Activity.fillna(last_mode,inplace=True)
```

```python
#find the mean of TotalVisits in df
visits_mean=df.TotalVisits.mean()
visits_mean
# fill the missing values with mode value of occupation in df.
df.TotalVisits.fillna(visits_mean,inplace=True)
```

```python
#find the mean of Page_Views_Per_Visit in df
page_visits_mean=df.Page_Views_Per_Visit.mean()
page_visits_mean
# fill the missing values with mode value of occupation in df.
df.Page_Views_Per_Visit.fillna(page_visits_mean,inplace=True)
```

```python
100*(df.isnull().sum()/len(df))
```

```python
df.head()
```

#For categorical variables with multiple levels, create dummy features (one-hot encoded)

```python
# Creating a dummy variable for some of the categorical variables and dropping the first one.
dummy1 = pd.get_dummies(df[['Lead Origin', 'Lead_Source', 'Last_Activity', 'occupation','choosing_a_course','Last Notable Activity']], drop_first=True)

# Adding the results to the master dataframe
df= pd.concat([df, dummy1], axis=1)
```

```python
df.shape
```

#Dropping the repeated variables

```python
df.drop(['Lead Origin','Prospect ID', 'Lead_Source', 'Last_Activity', 'occupation','choosing_a_course','Last Notable Activity'],axis=1,inplace=True)
```

```python
df.head()
```

#Step 4: Test-Train Split

```python
X=df.drop("Converted",axis=1)
y=df["Converted"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
print("Train X shape:",X_train.shape)
print("Test X shape:",X_test.shape)
print("Train y shape:",y_train.shape)
print("Test y shape:",y_test.shape)
```

#Step 5: Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
```

```python
from sklearn.feature_selection import RFE#Recursive Feature Elimination
from sklearn.linear_model import LogisticRegression

estimator=LogisticRegression()
selector=RFE(estimator,n_features_to_select=10)
selector.fit(X_train_scaled,y_train)
selector.support_
```

```python
cols_to_keep=X_train_scaled.columns[selector.support_]
```

```python
cols_to_keep
```

```python
X_train=X_train_scaled[cols_to_keep]
X_test=X_test_scaled[cols_to_keep]
```

#Step 6: Model Building

#first model

```python
import statsmodels.api as sm
```

```python
X_train_sm=sm.add_constant(X_train,has_constant="add")
X_test_sm=sm.add_constant(X_test,has_constant="add")
```

```python
model1=sm.GLM(np.array(y_train),X_train_sm,family=sm.families.Binomial())
# Binomial --> Binary Classification
result1=model1.fit()
result1.summary()
#Generalised Linear Model
```

```python
X_train_sm=X_train_sm.drop(["occupation_Housewife"],axis=1)
X_test_sm=X_test_sm.drop(["occupation_Housewife"],axis=1)
```

#second model

```python
model2=sm.GLM(np.array(y_train),X_train_sm,family=sm.families.Binomial())
result2=model2.fit()
result2.summary()
```

```python
X_train_sm=X_train_sm.drop(["Last Notable Activity_Had a Phone Conversation"],axis=1)
X_test_sm=X_test_sm.drop(["Last Notable Activity_Had a Phone Conversation"],axis=1)
```

#thrid model

```python
model3=sm.GLM(np.array(y_train),X_train_sm,family=sm.families.Binomial())
result3=model3.fit()
result3.summary()
```

```python
# Getting the predicted values on the train set
y_train_pred = result3.predict(X_train_sm)
y_train_pred[:10]
```

```python
y_train_pred = y_train_pred.values.reshape(-1)
y_train_pred[:10]
```

#Creating a dataframe with the actual converted  and the predicted probabilities

```python
y_train_pred_final = pd.DataFrame({'converted':y_train.values, 'converted_prob':y_train_pred})
y_train_pred_final['ID'] = y_train.index
y_train_pred_final.head()
```

#Creating new column 'predicted' with 1 if converted_prob > 0.5 else 0

```python
y_train_pred_final['predicted'] = y_train_pred_final.converted_prob.map(lambda x: 1 if x > 0.5 else 0)

# Let's see the head
y_train_pred_final.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=pd.DataFrame()
vif_data["Feature"]=X_train_sm.columns

vif_data["VIF"]=[variance_inflation_factor(X_train_sm.values,i) \
                 for i in range(len(X_train_sm.columns))]
vif_data
```

```python
train=pd.DataFrame(y_pred,columns=["Predicted_prob"])
train["Actual"]=y_train
train["Predicted_Class"]=np.where(train["Predicted_prob"]>0.5,1,0)
train=train.dropna()
```

```python
from sklearn import metrics
print("Accuracy =", metrics.accuracy_score(train["Actual"], train["Predicted_Class"]))
print("Recall =", metrics.recall_score(train["Actual"], train["Predicted_Class"]))
print("Precision =", metrics.precision_score(train["Actual"], train["Predicted_Class"]))
```
